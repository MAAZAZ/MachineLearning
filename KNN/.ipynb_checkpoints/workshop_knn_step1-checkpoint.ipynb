# Exercice 2

```python
le = preprocessing.LabelEncoder()

df=pd.read_csv("Iris.csv")
X=df[["SepalLengthCm", "SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]]
le.fit(X.loc[:,("Species")])
X.loc[:,("Species")]=le.transform(X.loc[:,("Species")])
X=np.array(X)
```

```python
le.classes_
```

```python
def normalisation(x,x_min,x_max):
    return (x-x_min)/(x_max-x_min)

def standarisation(x,u,v):
    return (x-u)/v
```

```python
 # La distance euclidienne entre 2 vecteurs
def distance_euclidien(v1, v2):
  distance = 0
  for i in range(len(v1)-1):
    distance += (v1[i]-v2[i])**2
  return sqrt(distance)

def getDistance(element):
  return element[-1]

# Recherche les voisins en fonction de K
def kvoisins(features, newdata, K):
  distances = list()
  voisins = list()

  for element in features:
    dist = distance_euclidien(newdata, element)
    distances.append((element, dist))
  
  # trier les distances
  distances.sort(key=getDistance)
    
  for i in range(K):
    voisins.append(distances[i])
  #print("les ",K," plus proches : ",voisins)
  return voisins

#  predicition (classsifier) de la valeur obtenu
def classifier_newData(features,newdata, K):
  voisins = kvoisins(features, newdata, K)
  # selectionner la distance
  data = [element[0][-1] for element in voisins]
  classes_count=[0,0,0]
  for i in data:
        if i==0:
            classes_count[0]+=1
        elif i==1:
            classes_count[1]+=1
        else:
            classes_count[2]+=1
  return classes_count
 
def getClass(classes_count):
    max=0
    key,keyIndex=0,0
    print(classes_count)
    for i in classes_count:
        if max<i:
            max=i
            keyIndex=key
        key+=1
        
    if(keyIndex==0):
        return "Iris-setosa"
    elif (keyIndex==1):
        return "Iris-versicolor"
    else:
        return "Iris-virginica"     
 
predict = classifier_newData(X, [5.9,3,5.1,1.8], 5)
print('La prediction de classe est : %s.' % (getClass(predict)))
```

```python
X=df[["SepalLengthCm", "SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
le.fit(df["Species"])
y=le.transform(df["Species"])
x_train,x_test,y_train,y_test=model_selection.train_test_split(X,y,random_state=2,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier=KNeighborsClassifier(n_neighbors=11,p=2,metric='euclidean')
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)

#cm=confusion_matrix(y_test,y_pred)
#accuracy_score(y_test,y_pred)
```

```python
def plotData():
    accuracy_list=[]
    k=[ i+3 for i in range(18)]
    for i in k:
        classifier=KNeighborsClassifier(n_neighbors=i,p=2,metric='euclidean')
        classifier.fit(x_train,y_train)
        y_pred=classifier.predict(x_test)
        accuracy_list.append(accuracy_score(y_test,y_pred))
        
    return k, accuracy_list


k,accuracy_list=plotData()
plt.figure(figsize=(12,8)) 
plt.plot(k,accuracy_list,color='blue',linestyle='dashed', marker='o',markerfacecolor='red',markersize=10)
plt.title('Accuracy vs K value')
plt.xlabel('K')
plt.ylabel('Accuracy')
print("Max accuracy: ",max(accuracy_list),"at K=",accuracy_list.index(max(accuracy_list)))
```
