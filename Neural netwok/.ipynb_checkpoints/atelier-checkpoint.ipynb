```python
#Program2: 

# Container for the Mersenne Twister pseudo-random number generator.
rng = np.random.RandomState(0) 
# Return a sample (or samples) from the “standard normal” distribution.
X = rng.randn(300, 2) 
# Compute the truth value of x1 XOR x2, element-wise.
y = np.array(np.logical_xor(X[:, 0] > 0, X[:, 1] > 0), dtype=int)
# SVM MODEL
svm_model = SVC(gamma='auto') 
svm_model.fit(X, y) 
# figure
fig = plt.figure(figsize=(10,8)) 
fig = plot_decision_regions(X=X, y=y, clf=svm_model, legend=2) 
plt.title("SVM")
plt.show()
```

```python
model3 = Sequential()

hidden_layer=Dense(2, input_dim=2,kernel_initializer='ones',bias_initializer='zeros')
model3.add(hidden_layer)
output_layer=Dense(2, kernel_initializer='random_uniform',bias_initializer='ones')
model3.add(output_layer)
print(model3.summary())
print(model3.get_weights()) #initial weights
```

```python
model3.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model3.fit(X, y,epochs=10)
```

```python
fig = plt.figure(figsize=(10,8)) 
fig = plot_decision_regions(X=X, y=y, clf=model3, legend=2) 
plt.title("SVM")
plt.show()
```

```python
# Initializing Classifiers
clf1 = LogisticRegression(random_state=1, solver='lbfgs')
clf2 = RandomForestClassifier(n_estimators=100, 
                              random_state=1)
clf3 = GaussianNB()
clf4 = SVC(gamma='auto')

gs = gridspec.GridSpec(2, 2)

fig = plt.figure(figsize=(10,8))

labels = ['Logistic Regression', 'Random Forest', 'Naive Bayes', 'SVM']
for clf, lab, grd in zip([clf1, clf2, clf3, clf4],
                         labels,
                         itertools.product([0, 1], repeat=2)):

    clf.fit(X, y)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X, y=y, clf=clf, legend=2)
    plt.title(lab)

plt.show()
```
