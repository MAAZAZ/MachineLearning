```python
import matplotlib.pyplot as pt
import numpy as np
```

```python
def f(x):
    return x**2

x=np.arange(-10,10,0.1)
y=f(x)

#pt.plot(1,2,"ro")
#pt.plot(20,4,"ro")
#pt.plot([2,2],[4,5],"gx")

pt.plot(x,f(x))
pt.plot(-7,f(-7),"rx")
pt.xlabel("x")
pt.ylabel("y")
pt.legend(["x","(-7,f(7))"])
pt.axis([-0,1,0,1])
```

```python
def h(x):
    return np.cos(np.pi * x)
def g(x):
    return x*np.cos(np.pi * x)

x = np.arange(-1.0, 2.0, 0.01)

pt.plot(x,h(x),label = "cos(pi*x)")
pt.plot(x, g(x),linestyle='dashed', color='red',label = "x*cos(p*x)")
pt.legend()
```
