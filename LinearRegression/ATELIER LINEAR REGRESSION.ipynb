```python
########### exerice5 
#Q1
def learing2(X): 
    w0=0
    w1=0 
    nb_iterations=50 
    for i in range(nb_iterations):
        y_predicted=w1*X+w0
        mse=metrics.mean_squared_error(X,y_predicted)
        print("w0:{} --- w1:{} -- y_predicted:{} --> ERROR: {}\n".format(w0,w1,y_predicted,mse))
        w1=w1+0.4

learing2(np.array(X))
```

```python
#Q2
# Pourquoi la valeur minimale de MSE nest pas atteinte ?
# a cause de step utilis, nous avons saut la valeur minimale

#Q3
# Est-ce que le fait daugmenter le nombre ditration peut assurer lobtention de MSE minimale ?
# Non

#Q4
# Il existe deux problmes dans le code prcdent, lesquels ?
# le problme du choix de la premire itration de w1 et le step utilis pour ne pas sauter la valeur minimale

#Q5
# Comment la drive partielle peut aider  rsoudre lun des deux problmes ? Lequel ? (technique1)
# la drive partielle peut aider  rsoudre le problme de la direction car le graphe est sous forme d'un courbe hyperbolique
# la seule chose qui peut assurer la direction, c'est la drive partielle

#Q6
# Comment la constante learning_rate peut aider  rsoudre le deuxime problme ? lequel ?
# (le premier est rsolu  laide de gradient descent) (technique2)
# learning_rate nous permet de ne pas sauter la valeur minimale (les steps)

#Q7
# Modifier la fonction de manire  intgrer les deux techniques pour rsoudre les deux problmes.
def findW(x):
    w0=0
    w1=0
    n_iter=5000
    list_w1=[]
    list_mse=[]
    learning_rate=0.001
    for i in range(n_iter):
        y_pred=w0+w1*x
        mse=metrics.mean_squared_error(y,y_pred)
        list_w1.append(w1)
        list_mse.append(mse)
        dw1=(-2/len(x))*np.sum(x*(y-y_pred))
        w1=w1-learning_rate*dw1
        if mse == 0:
            return -1
    
    return list_w1,list_mse

findW(np.array(X))
# fonction ==> gradient descent
```

```python
pt.plot(findW(np.array(X))[0],findW(np.array(X))[1])
pt.xlabel("w1")
pt.ylabel("mse(w1)")
```
