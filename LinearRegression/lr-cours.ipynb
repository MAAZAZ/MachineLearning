```python
# mse en fct de w1 (w0=0)
X=ng.array([1,2,3,4])
y_real=ng.array([1,2,3,4])

def findW(x):
    w0=0
    w1=0
    n_iter=10000
    list_w1=[]
    list_mse=[]
    learning_rate=0.001
    for i in range(n_iter):
        y_pred=w0+w1*x
        mse=metrics.mean_squared_error(x,y_pred)
        list_w1.append(w1)
        list_mse.append(mse)
        dw1=(-2/len(x))*ng.sum(x*(y_real-y_pred))
        w1=w1-learning_rate*dw1
    
    return list_w1,list_mse

findW(X)
# fonction ==> gradient descent
```

```python
from matplotlib import pyplot as pt
```

```python
pt.plot(findW(X)[0],findW(X)[1])
pt.xlabel("w1")
pt.ylabel("mse(w1)")
```

```python
import numpy as ng
from matplotlib import pyplot as pt
# mse en fct de w1 (w0=0)
X=ng.array([1,2,3,4])
y_real=ng.array([1,2,3,4])

def findW(x):
    w0=0
    w1=0
    n_iter=100
    list_w1=[]
    list_mse=[]
    learning_rate=0.001
    for i in range(n_iter):
        y_pred=w0+w1*x
        mse=metrics.mean_squared_error(x,y_pred)
        list_w1.append(w0)
        list_mse.append(mse)
        dw0=(-2/len(x))*ng.sum(y_real-y_pred)
        w0=w0-learning_rate*dw0
    
    return list_w1,list_mse

findW(X)
# fonction ==> gradient descent
```

```python
pt.plot(findW(X)[0],findW(X)[1])
pt.xlabel("w0")
pt.ylabel("mse(w0)")
```
